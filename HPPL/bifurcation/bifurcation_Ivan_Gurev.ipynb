```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline
```

```python
def step(r,x):
    return r*x*(1-x)
```

```python
x = np.linspace(0,1,1000)
r = 2
plt.plot(x,step(r,x))
plt.xlabel('$x_n$')
plt.ylabel('$x_{n+1}$')
plt.title(f'r = {r}');
```

```python
def iterat_r(x0, r_max, m, n, steps):
    R = np.linspace(0,r_max, steps)
    X = list()
    R_list = list()
    for r in R:
        x = x0
        for i in range(m+n):
            if (i>n) and (x not in X):
                X.append(x)
                R_list.append(r)
            x = step(r,x)
    return X, R_list
```

```python
X, R = iterat_r(0.1,4,100, 200,200)
```

```python
plt.plot(R, X, ls='', marker=',')
plt.xlabel('r')
plt.ylabel('x');
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, 4.1), ylim=(0, 2))
line, = ax.plot([], [], ls='', marker=',')

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = R[:i]
    y = X[:i]
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=len(R), interval=2, blit=True)



anim.save('test.gif', writer='Pillow')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Poison_flow():
    
    _Nl = 9   #Number of directions
    
    #vectors f_i are counted from top in clockwise direction, e.g f1=12,f2 = 1.5, f3=3, f5=6 etc.
    _xs = np.array([0, 0, 1, 1, 1, 0,-1,-1,-1])
    _ys = np.array([0, 1, 1, 0,-1,-1,-1, 0, 1])
    _w = np.array([4/9,1/9,1/36,1/9,1/36,1/9,1/36,1/9,1/36])
    
    pin = 1.5
    pout = 0.5
    idx = np.arange(_Nl)
    plotRealTime = True
    
    def __init__(self, Nx, Ny, tau):
        self.Nx = Nx
        self.Ny = Ny
        self.tau = tau
            
    
    def get_F(self):
        
        F = np.ones((self.Ny,self.Nx,self._Nl))/9.0
            
        return F
    
 
    def get_fbound(self,F):
        #left bound
        ux = -(F[:,0,0]+F[:,0,1]+F[:,0,5]+2*(F[:,0,6]+F[:,0,7]+F[:,0,8]))/self.pin + 1
        F[:,0,3] = F[:,0,7]+2/3*self.pin*ux
        F[:,0,2] = F[:,0,6]+0.5*(F[:,0,5]-F[:,0,1])+1/6*self.pin*ux
        F[:,0,4] = F[:,0,8]+0.5*(F[:,0,1]-F[:,0,5])+1/6*self.pin*ux
        
        #lower bound
        F[0,:,1] = F[0,:,5]
        F[0,:,2] = F[0,:,6]+0.5*(F[0,:,7]-F[0,:,3])
        F[0,:,8] = F[0,:,4]+0.5*(F[0,:,3]-F[0,:,7])
        
        #right bound
        #ux = -(F[:,-1,0]+2*F[:,-1,1]+F[:,-1,2]+2*F[:,-1,5]+F[:,-1,4]+2*F[:,-1,8])/self.pout + 1
        #F[:,-1,7] = F[:,-1,3]-2/3*self.pout*ux
        #F[:,-1,8] = F[:,-1,4]+0.5*(F[:,-1,5]-F[:,-1,1])+5/3*self.pout*ux
        #F[:,-1,6] = F[:,-1,2]+0.5*(F[:,-1,1]-F[:,-1,5])+5/3*self.pout*ux
        
        #upper bound
        F[-1,:,5] = F[-1,:,1]
        F[-1,:,4] = F[-1,:,8]+0.5*(F[-1,:,7]-F[-1,:,3])
        F[-1,:,6] = F[-1,:,2]+0.5*(F[-1,:,3]-F[-1,:,7])

        
        #Top Left bound
        F[-1,0,3] = F[-1,0,7]
        F[-1,0,4] = F[-1,0,8]
        F[-1,0,5] = F[-1,0,1]
        F[-1,0,2] = F[-1,0,6] = 0.5*(self.pin-(F[-1,0,0]+F[-1,0,1]+F[-1,0,3]+F[-1,0,4]+F[-1,0,5]+F[-1,0,7]+F[-1,0,8]))
        
        #Top Right bound
        #F[-1,-1,6] = F[-1,-1,2]
        #F[-1,-1,5] = F[-1,-1,1]
        #F[-1,-1,7] = F[-1,-1,3]
        #F[-1,-1,8] = F[-1,-1,4] = 0.5*(self.pout-(F[-1,-1,0]+F[-1,-1,1]+F[-1,-1,2]+F[-1,-1,3]+F[-1,-1,5]+F[-1,-1,6]+F[-1,-1,7]))
        
        #Bottom Left bound
        F[0,0,2] = F[0,0,6]
        F[0,0,1] = F[0,0,5]
        F[0,0,3] = F[0,0,7]
        F[0,0,8] = F[0,0,4] = 0.5*(self.pin-(F[0,0,0]+F[0,0,1]+F[0,0,2]+F[0,0,3]+F[0,0,5]+F[0,0,6]+F[0,0,7]))
          
        #Bottom Right bound
        #F[0,-1,1] = F[0,-1,5]
        #F[0,-1,7] = F[0,-1,3]
        #F[0,-1,8] = F[0,-1,4]
        #F[0,-1,2] = F[0,-1,6] = 0.5*(self.pout-(F[0,-1,0]+F[0,-1,1]+F[0,-1,5]+F[0,-1,3]+F[0,-1,4]+F[0,-1,7]+F[0,-1,8]))
        
        
        return F
        
    def simulate(self,Nt):
        
        F = self.get_F()
                    
        fig = plt.figure(figsize=(4,2),dpi=80)
        
        for it in range(Nt):
    
            for i,cx,cy in zip(self.idx,self._xs,self._ys):
                F[:,:,i] = np.roll(F[:,:,i], cx, axis=1)
                F[:,:,i] = np.roll(F[:,:,i], cy, axis=0)
            F = self.get_fbound(F)
            
            rho = np.sum(F,2)
            ux = np.sum(F*self._xs,2) / rho
            uy = np.sum(F*self._ys,2) / rho
            
            
            
            Feq = np.zeros(F.shape)
            for i, cx, cy, w in zip(self.idx,self._xs,self._ys,self._w):
                Feq[:,:,i] = rho*w*(1+3*(cx*ux+cy*uy)+9*(cx*ux+cy*uy)**2/2-3*(ux**2+uy**2)/2)
             
            F += -(1.0/self.tau)*(F-Feq)
            
            
            if (self.plotRealTime and (it%25)==0) or (it == Nt-1):
                plt.imshow(np.sqrt(ux**2+uy**2))
                ax = plt.gca()
                plt.pause(0.001) 
                
                if it==Nt-1:
                    plt.plot(np.arange(self.Ny),ux[:,self.Nx//2])
                    plt.xlabel('y')
                    plt.ylabel('$u_x$')
            
```

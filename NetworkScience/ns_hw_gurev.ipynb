## Problem 1A
Consider an undirected network in which each node has the clustering coefficient $C_{i} = 0$ and
degree $k = 2$.

a. What is the minimal number of nodes $N$ and edges $E$ in this network? Draw such a network

b.  How many disconnected components can a network with these properties possibly have?

## Solution

### a

1. Degree $k=2$ means that each node is connected to exactly 2 other nodes. this implies that the network consists of a series of interconnected nodes forming a chain or a loop.

2. The clustering coefficient $C_i = 0$ for each node indicates that there are no triangles in the network. In other words, the network does not contain any closed loops or cycles of length 3.

Given these conditions, let's consider the minimal case:

In order to satisfy $k=2$ the network has to consist of at least 3 nodes forming a triangle. To satisfy $C_i = 0$ the network has to be a square. As a result, the minimal number of nodes $N=4$ and edges $E=4$. 

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4])

G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

### b

With such properties a network can have $\text{Comps} = int(\frac{N}{4})$

## Problem 2A

probability $p = 10^{-3}$.

a. What is the expected number of edges $\langle E\rangle$ in such a network?

b. In which regime is the network?

c. Estimate the link probability $p_c$ at which the network is ath the critical point (percolation).

d. What is the scaling of the size of the giant component at the critical point with $N$?

### a

$$ \langle E \rangle = \binom{N}{2}p = \frac{N(N-1)}{2}p = 4498.5 \approx 4499

### b
$$\langle k\rangle = p(N-1) = 2.9 > 1$$

The network is in supercritical regime

### c
$$p_c \approx \frac{1}{N} = 3.33\cdot 10^{-4}$$

### d
At critical point:

$$N_G \sim N^{2/3} = 208 $$

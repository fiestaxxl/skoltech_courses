### Problem 1.0 (Matrix norms) (8 pts)

1. For any norm $\|\cdot\|$ in finite dimentional space $\mathbb{C}^{N}$ we define dual norm $\|x\|^{'} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\}$, where ${\cdot}^{\star}$ denotes transpose complex-conjugate vector.

+ (2 pts). Prove that $\|\cdot\|^{'}$ is a norm.

+ (2 pts). Find dual norms for Chebyshev norm $\left\|x\right\|_{\infty} = \max\limits_{i=1,\dots, N}|x_{i}|$ and $l_1$ norm $\left\|x\right\|_1 = \sum\limits_{i=1}^{N}|x_{i}|$.

+ (2 pts). Find dual norm for $l_p$ norm $\left\|x\right\|_{p} = \left(\sum\limits_{i=1}^{N}|x_{i}|^{p}\right)^{1/p}$.

+ (2 pts). Find dual norm for weighted norm $\left\|x\right\|_{A} = \left(x^T A x\right)^{1/2}$ with positive-definite matrix $A$ and $x\in\mathbb{R}^{N}$.

### Solution

#### 1. 

$\|x\|^{'}$ is a norm if:
+ $\|\alpha x\|^{'} = |\alpha|\|x\|^{'}$

*proof:*
$\|\alpha x\|^{'} = \sup\left\{\left|\alpha y^{\star} x\right|:\left\|y\right\|=1\right\} = \sup\left\{\left|\alpha\right|\left| y^{\star} x\right|:\left\|y\right\|=1\right\} = \left|\alpha\right|\sup\left\{\left| y^{\star} x\right|:\left\|y\right\|=1\right\} = \left|\alpha\right|\| x\|^{'}$

+ $\| x + y\|^{'} \leq \| x \|^{'}+ \|y\|^{'}$

*proof:*
$\| x + z\|^{'} = \sup\left\{\left| y^{\star} (x+z)\right|:\left\|y\right\|=1\right\} \leq \sup\left\{\left| y^{\star} x\right|+\left| y^{\star} z\right|:\left\|y\right\|=1\right\} \leq \sup\left\{\left| y^{\star} x\right|:\left\|y\right\|=1\right\} + \sup\left\{\left| y^{\star} z\right|:\left\|y\right\|=1\right\} = \| x \|^{'}+ \|z\|^{'}$

+ if $\| x \|^{'} = 0 $ then $x=0$

*proof:*
It is clear, that  $\| 0 \|^{'} = 0 $. If $x \neq 0$, then $\| x \| \neq 0 $ and taking $y=\frac{x}{\| x \|}$ we have $\| x \|^{'}  \geq \frac{\| x \|^{2}_2}{\| x \|} > 0 $. So, if $\| x \|^{'} = 0$, $x$ must be equal to 0 

#### 2.
Chebyshev norm:

$\| x \|^{'}_\infty = \sup\left\{\left|y^{\star} x\right|:\left\|y \right\|_\infty=1\right\}$


$| y^{\star}x | = | \sum^{n}_{i=1}\bar y_i x_i |\leq \sum^{n}_{i=1}|\bar y_i x_i| \leq \max{|\bar y_i|}\cdot \sum^{n}_{i=1}|x_i| = \left\|y \right\|_\infty \cdot \left\|x \right\|_1 $

$\| x \|^{'}_\infty = \sup\left\{\left|y^{\star} x\right|:\left\|y \right\|_\infty=1\right\} = \left\|x \right\|_1$

$l_1$ norm:

$\| x \|^{'}_1 = \sup\left\{\left|y^{\star} x\right|:\left\|y \right\|_1=1\right\}$


$| y^{\star}x | = | \sum^{n}_{i=1}\bar y_i x_i |\leq \sum^{n}_{i=1}|\bar y_i x_i| \leq \max{| x_i|}\cdot \sum^{n}_{i=1}\bar|y_i| = \left\|y \right\|_1 \cdot \left\|x \right\|_\infty $

$\| x \|^{'}_1 = \sup\left\{\left|y^{\star} x\right|:\left\|y \right\|_1=1\right\} = \left\|x \right\|_\infty$

#### 3.

$\| x \|^{'}_p = \sup\left\{\left|y^{\star} x\right|:\left\|y \right\|_p=1\right\}$

$| y^{\star}x | =| \sum^{n}_{i=1}\bar y_i x_i |\leq \sum^{n}_{i=1}|\bar y_i x_i| = \sum^{n}_{i=1}|\bar y_i|| x_i| \leq (\sum^{n}_{i=1}|x_i|^q)^{1/q}(\sum^{n}_{i=1}|\bar y_i|^p)^{1/p} = \| x \|_q\| y \|_p, 1/q + 1/p = 1$ - from Holder's inequality

$\| x \|^{'}_p = \sup\left\{\left|y^{\star} x\right|:\left\|y \right\|_p=1\right\} = \| x \|_q $

#### 4.

Under assumtion A is positive defined and symmetric, A can be decomposed:

$$ A = R R^T $$

where $R$ - lower triangular matrix

$$ z = R^Ty$$
$$y^TAy = y^TRR^Ty = z^Tz$$

As a result:

$\| x \|^{'}_A = \sup\left\{\left|y^T x\right|:\left\|y \right\|_A=1\right\} = 
 \sup\left\{\left|x^T R^{-T}z\right|:\left\|z \right\|_2=1\right\} = 
 \sup\left\{\left|z^T R^{-1}x\right|:\left\|z \right\|_2=1\right\} = 
 \| R^{-1}x\|^{'}_2 = \|R^{-1}x\|_2 = (|x^TR^{-T}R^{-1}x|)^{\frac{1}{2}} = (x^TA^{-1}x)^{\frac{1}{2}} = \|x\|_{A^{-1}}$

### Problem 1.1 (Matrix norms) (10 pts)

1. Let $U \in \mathbb{R}^{n \times r}$ be an $n \times r$ matrix with orthonormal columns $U^{\top} U = I_r$. Let $V$ be an $m \times r$ matrix. 

+ (2 pts). Propose an algorithm for the computation of $\Vert A \Vert^2_F$, where $A = U V^{\top}$, and estimate its complexity with respect to $n, m, r$. How orthogonalization can be used? 

+ (3 pts). Consider the matrix $B = A \circ A$, where $\circ$ is the elementwise product of matrices (i.e., the elements of the matrix $B$ are squares of the elements of the matrix $A$). What is the maximal possible rank of the matrix $B$? 

+ (5 pts). Propose an algorithm for the computation of $\Vert B \Vert^2_F$ and estimate its complexity with respect to $n, m, r$.

### Solution

#### 1.
$\| A \|^{2}_F = tr(A^TA) = tr((U V^T)^T(UV^T)) = tr((V^T)^T(U^TU)(V^T)) = tr(VV^T) = \|V\|^{2}_F$

Time complexity: $T(n,m,r) = O(m \cdot \min(r,n)) $

#### 2.
Let $U$, $V$ be full-rank matrices. Then 

$$
rank(A) \leq rank(U)rank(V^T) = \min(n,r)\min(m,r)
$$

There exists matrix A for which equality holds. Let $d = \min(n,r)\min(m,r)$. Then maximum rank of the matrix $B = A \circ A$ is $\binom{d+1}{2} = \frac{d(d+1)}{2}$ 

*proof:*

Writing $A$ as a sum of rank-one matrices:

$$
A = \sum_{i=1}^{d}p_iq_i^T
$$

results in

$$
A \circ A = \sum_{i=1}^d\sum_{j=1}^d(p_i \circ p_j)(q_i \circ q_j)^T
$$

By finding the max number of distinct rank-one matrices, we can see, that it follow simple combinatorics:

$$
rank(A \circ A) \leq \binom{d+2-1}{2} = \binom{d+1}{2}
$$

#### 3.

$$A = UV^T = \sum_{i=1}^ru_iv^T_i.$$

$$ B = A \circ A = (\sum_{i=1}^ru_iv^T_i)\circ(\sum_{j=1}^ru_jv^T_j) = \sum_{i=1}^r \sum_{j=1}^r(u_iv^T_i)\circ(u_jv^T_j) = \sum_{i=1}^r \sum_{j=1}^r(u_i\circ u_j)(v_i \circ v_j)^T$$

$$\| B \|^2_F = Tr(BB^T) = Tr(\sum_{i=1}^r \sum_{j=1}^r(u_i\circ u_j)(v_i \circ v_j)^T\sum_{k=1}^r \sum_{l=1}^r(v_k\circ v_l)(u_k \circ u_l)^T) = \sum_{i=1}^r \sum_{j=1}^r\sum_{l=1}^r \sum_{k=1}^r(v_i \circ v_j)^T(v_k\circ v_l)Tr((u_i\circ u_j)(u_k \circ u_l)^T) = \sum_{i=1}^r \sum_{j=1}^r\sum_{l=1}^r \sum_{k=1}^r(v_i \circ v_j)^T(v_k\circ v_l)(u_k\circ u_l)^T(u_i \circ u_j).$$

Computin in sum term is $O(n+m)$. As a result, total complexity is $O((n+m)r^4)$

### Problem 1.2 (Unitary and orthogonal matrices) (4 pts)

+ (2 pts). Let $O\in\mathbb{R}^{n\times n}$ be orthogonal matrix. Characterise explicitly all orthogonal matrices that are positive definite. You may start with $n=2$.

+ (2 pts). Show that any unitary matrix from $\mathbb{C}^{n\times n}$ can be represented as a product of at most $n$ Householder reflectors.

### Solution

#### 1.

Let's consider a set $R$, which is intersection of set of orthogonal matrices and set of positive definite matrices. Identity matrix $I$ is both positive definite and orthogonal that is the set $R$ is non-empty 

Any orthogonal $n$-by-$n$ matrix $Q$ can be produced by Cayley transform:

$$
Q = (I-A)(I+A)^{-1}
$$

where $A$ - anti-symmetric matrix ($A^T = -A$), for which $I+A$ is invertible.

Furthermore, if $Q$ doen't have -1 as eigenvalue, then:

$$
A = (I-Q)(I+Q)^{-1}
$$

This provides a rational parametrization of the orthogonal group minus set consisting of the orthogonal matrices that have -1 as eigenvalue.

As a result, $Q$ satisfies $x^TQx>0$ for all nonzero $x \in {\rm I\!R} ^n$ iff:

$$
0 < x^TQx = y^T(I+A)(I-A)y = y^T(I-A^2)y = |y|^2 - |Ay|^2
$$

where $x=(I+A)y$

The equation above is the equivalent of the fact that $\|A\| < 1$. Thus, set $R$ is parametrized by the open convex set in the vector space of anti-symmetric $n$-by-$n$ matrices consisting of those anti-symmetric matrices $A$ whose matrix norm is less than 1.

#### 2.
To proove the statement let's consider real case first and then expand it to the complex one

By using $n-1$ Householder transforms $H_i = I_n - 2v_iv_i^T$ we can upper triangularize given orthogonal U as:

$$
H_{n-1}H_{n-2}...H_2H_1U = R
$$

It should be noticed that $R$ is both upper triangular and orthogonal. Looking at the $I_n = R^TR$ establishes that $R$ is diagonal with diagonal elements equal to $\pm 1$. For euclidian inner product case it could be seen that $r_{11}=...=r_{n-1,n-1} = 1$ can be achieved.
In case the last diagonal entry $r_{nn}$ turns out to be $-1$, we set $H_n := I_n - 2e_ne_n^T$ and obtain the following:

$$
H_nR = H_nH_{n-1}...H_1U = I_n
$$

As each Householder matrix $H_i$ is its own inverse, we have expressed:

$$
U = H_1...H_n
$$

as the product of $n$ Householder matrices if $r_{nn} = -1$ in $R$.

Expanding to complex unitary case we should state, that $|r_{ii}| = 1$ in $R$. Specifically, $r_{ii} = 1$ can be achieved for all $i<n$ with judicious choices for the individual complex Householder transforms $H_i$

1. We define approximate $\epsilon-$rank of matrix $A$ as $\text{rank}_{\epsilon}A = \min \left\{\text{rank}(X): \left\|A-X\right\|_{\max}\leq \epsilon\right\}$.

  
+ (6 pts). Let $f(x, y),\,x,y\in[0, 1]$ is analytic function with $\left.\partial_y^{(k)}f(x, y)\right|_{y=0}\leq M$ for all $k$. Show that approximate rank of matrix $A_{ij} = f(x_i, y_j)$ where $x_{i},\,i=1,\dots,p$, $y_{j},\,j=1,\dots,q$ are arbitrary points inside $[0, 1]^2$, is bounded independent of $p$, $q$ and the choice of points.

+ (6 pts). Extend result from the previous point (a.) to $x, y \in \mathbb{R}^{N}$. Observe how the bound depends on $N$.

+ (3 pts). Demonstarate your bounds from two previous points numerically on functions

    + (1 pts). $\cos(xy)$, $\exp(-(x-y)^2)$ and a few on your choice for $N=1$
    + (1 pts). [RBF kernel](https://scikit-learn.org/stable/modules/gaussian_process.html#radial-basis-function-rbf-kernel) for different $N$.
    + (1 pts). [Exp-Sine-Squared kernel](https://scikit-learn.org/stable/modules/gaussian_process.html#exp-sine-squared-kernel) for different $N$.


### Solution

#### 1.
$$f(x,y) \approx \sum_{k=0}^n\bigg(\frac{y^k}{k!}\frac{\partial^{(k)}{f}}{\partial{y}^k}\bigg) + \frac{y^{(n+1)}}{(n+1)!}\frac{\partial^{(n+1)}{f}}{\partial{y}^{(n+1)}} $$

$$ |f(x,y) - \sum_{k=0}^n\bigg(\frac{y^k}{k!}\frac{\partial^{(k)}{f}}{\partial{y}^k}\bigg)| \leq \frac{y^{(n+1)}}{(n+1)!}M \leq \frac{M}{(n+1)!}, y \in [0,1]$$ 

As a result, matrix $X$ can be set to be:

$$
X_{ij} = \sum_{k=0}^n \frac{y_j^k}{k!} \frac{\partial^{(k)}{f(x_i,y_j=0)}}{\partial{y}^k}
$$

It can be seen, that every column in $X$ is:

$$
X_j = \left(\begin{matrix}
\sum_{k=0}^n\frac{y^k_j}{k!}\partial_y^{(k)}f(x_1,0) \\ \sum_{k=0}^n\frac{y^k_j}{k!}\partial_y^{(k)}f(x_2,0) \\ \vdots \\ \sum_{k=0}^n\frac{y^k_j}{k!}\partial_y^{(k)}f(x_p,0)
\end{matrix}\right) 
= \sum_{k=0}^n\frac{y^k_j}{k!}\left(
\begin{matrix} \partial_y^{(k)}f(x_1,0) \\ \partial_y^{(k)}f(x_2,0) \\ \vdots \\ \partial_y^{(k)}f(x_p,0)
\end{matrix}\right),$$


As a result, $$rank(X) = n$$

Let's estimate $\min(rank(X))$:

$$
\frac{M}{(n+1)!} \leq \epsilon
$$

$$
\frac{M}{\epsilon} \leq (n+1)!
$$

$$
rank(X) = \min(n \in \mathbb{N}: (n+1)! \geq \frac{M}{\epsilon})
$$

+ (4 pts). Use the result from the previous point to construct an algorithm that distinguishes different molecules. Implement it and run on the database. Produce a list of distinct molecules and visualise them.

```python
def getDistinct(molecules):
    
    x = np.shape(molecules)
    Norm = np.empty((x[0],2))
    basis = np.empty((x[0],x[1]-1,x[2]))
    
    for i in range(x[0]):
        for j in range(x[1]-1):
            if j != x[1]-2:
                basis[i,j,:] = molecules[i,j+2] - molecules[i,j+1,:]
            else: 
                basis[i,j,:] = molecules[i,1,:] - molecules[i,x[1]-1,:]
        Norm[i,0] = i
        Norm[i,1] = np.round(np.linalg.norm(basis[i],2),7)
    
    return basis, Norm
```

```python
X, Y = getDistinct(molecules)

unique_mol = []
j = 0
i = 0
step = 0
Z = sorted(np.unique(Y[:,1]),reverse = True )
while i < len(Z):
    norm = Z[j]
    if norm == Y[step,1]:
        unique_mol.append(step)
        i += 1
        j += 1
    else:
        step += 1
```

+ (7 pts). Learn about [Kabsch algorithm](https://en.wikipedia.org/wiki/Kabsch_algorithm). Implement it and run on the database. Produce a list of distinct molecules and visualise them.

```python
def rotation_kabsch(P,Q):
    
    H = np.dot(np.transpose(P),Q)
    u,s,v = np.linalg.svd(H)
    d = np.sign(np.linalg.det(np.dot(v,u.T)))
    S = np.diag([1,d])
    R = np.dot(np.dot(v,S),u.T)
    
    return R, mse(Q,np.dot(R,P.T).T)
```

```python
i = 0
j = 1
unique = np.arange(len(molecules))
eps = 1e-6
mol = molecules[0]

while True:
    nonunique = list()
    
    for k in range(j,len(molecules)):
        M = molecules[k]
        R, err = rotation_kabsch(mol,M)
        if err < eps:
            nonunique.append(k)
    unique = unique[~np.in1d(unique,nonunique)]
    if unique[i] == unique[-1]:
        break
        
    i += 1
    mol = molecules[unique[i]]
    j = unique[i+1]
print('Molecules',*unique,'are unique' )
```

```python
fig, ax = plt.subplots(1, len(unique), figsize=(12, 6))
ax = ax.reshape(-1,)
for ind, mol in enumerate(unique):
    draw_molecule(ax[ind], molecules[mol])
```

## Brief intro

The main component of [GAN](https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html) is Generator network. Generator is such model that takes as input random vector and produces an image. In this task, we will work with the already pre-trained generator $G$. Here are a few things you need to know about this model:



1.   **Model input:** batch of random vectors $z\sim N(0,1)$ with normally distributed values of size **(Batch_Size,100)**. Here, Batch_Size is a number of images we want to generate
2.   **Model output:** $\mathrm{images} = G(z)$ - batch of images with size **(Batch_Size,3,64,64)**. Here **3** - defines number of image channels (3 for RGB images), **64, 64** - width and height.
3.   **Model structure:** generator is a set of sequentially appplied functions. Each function input is the output of the previous one. In other words, $G(z) = f_N(f_{N-1}...f_1(f_0(z)))$. $N$ - defines number of functions in generator.



### Disentanglement

The good property of GAN is that generator input space has a set of meaningful directions - moving input vector along this directions produce interpretable changes in image. For example, we can change a Pose, Age, Expression or even add eyeglasses. You can read more about it on [resource](https://genforce.github.io/interfacegan/). The main question is how to find such good directions. This problem is called disentanglement.

![](https://github.com/genforce/interfacegan/blob/master/docs/assets/teaser.jpg?raw=true)

## Problem discription 

We are going to find meaningful directions using SVD applied over the Jacobian matrix of some hidden output with respect to model input. **i-th** hidden output of generator model is defined by formula $G_{i}(z) = f_i(f_{i-1}...f_1(f_0(z))))$, where $i < N$. So, the task could be splitted on four main parts:

1. **(4 pts)**   Write the function **Get_Hidden_Output**, which reproduces formula $G_{i}(z) = f_i(f_{i-1}...f_1(f_0(z))))$
2. **(3 pts)**   Calculate jacobian matrix $J$ of generator $G$ for several **i** values. For each of them:

*   Calculate the jacobian matrix with respect to $z$: $\frac{\partial G(z)}{\partial z}$
*   Reshape jacobian matrix to have the size **(Batch_Size * Hidden_Output_Size,100)**

3. **(3 pts)**  Apply SVD to jacobian matrix ($J = U @ S @ V.T$) from previous step

4. **(2 pts)**  Take several directions from matrix $V$ (particular columns), and run visualization code (already written for you). Leave a comment on what you got (how found directions affect image)

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )
    def get_function_output(self, inp, index=0):
      if index == 0:
        inp = inp[:,:,None,None]
      return self.main[index](inp)
    def get_number_of_layers(self):
      return len(self.main)

    def forward(self, input):
        return self.main(input[:,:,None,None])
```

**Define the function that returns the output of any generator function for given input z and generator function index (4 pts)**

```python
def Get_Hidden_Output(z,index):
  out = z
  for i in range(index):
    out = G.get_function_output(out,index)
  return out
```

**Define the function that calculates the jacobian matrix based on given input z and generator function index (use the function which you implemented earlier) (3 pts)**

```python
def Calc_Gradient(z, index):
  func = partial(Get_Hidden_Output, index = index)
  jacobian = jacfwd(func)
  grads = vmap(jacobian)(z.unsqueeze(1))
  return grads
```

**Calculate SVD and visualize singular values for several generator function indexes (2-3 indexes are enough). Based on singular value plots, decide what generator function is more useful for disentangled representation search and why, leave a comment.(3 pts)**

```python
index = [2,3,4]
z = sample_batch(20)
V_list = list()
S_list = list()
for ind in index:
    grads = Calc_Gradient(z,ind)
    grads = grads.reshape(-1,100)
    U, S, V = torch.linalg.svd(grads, full_matrices=False)
    S_list.append(S)
    V_list.append(V.T)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
for ind, val in enumerate(S_list):
  ax.scatter(range(np.shape(val)[0]),val.cpu().detach().numpy(), label = f'index = {index[ind]}')
plt.xlabel('index')
plt.ylabel('$\sigma$')
plt.legend()
```

The larger the values of $\sigma$ the more information image contains.As a result, generator functions with higher $\sigma$ give us better adjustments in the final image.
